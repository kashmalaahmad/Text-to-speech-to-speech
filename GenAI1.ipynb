```python
import pdfplumber
import requests
import pydub
import os
```

```python
def pull_text_from_pdf(pdf_path):
    all_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text() or ""
            all_text += page_text
            print(f"Grabbed from page {page.page_number}: {page_text[:100]}...")
   
    all_text = " ".join(all_text.split())
    print(f"Total text grabbed: {len(all_text)} characters")
    return all_text
```

```python
pdf_path = "/Life 3.0-Being Human in the Age of Artificial Intelligence.pdf"
grabbed_text = pull_text_from_pdf(pdf_path)
print("Sneak peek of text:", grabbed_text[:500])
```

```python
def split_into_chunks(text, max_chars=500):
    chunks = [text[i:i + max_chars] for i in range(0, len(text), max_chars)]
    print(f"Made {len(chunks)} chunks")
    return chunks
```

```python
text_pieces = split_into_chunks(grabbed_text)
print("First chunk peek:", text_pieces[0] if text_pieces else "No text")
```

```python
from gtts import gTTS
import os
from pydub import AudioSegment

def make_speech_from_chunks(text_chunks):
    audio_bits = []
    for idx, chunk in enumerate(text_chunks):
        tts = gTTS(text=chunk, lang='en')
        temp_file = f"temp_{idx}.mp3"
        tts.save(temp_file)
        audio_bits.append(temp_file)
        print(f"Made speech for chunk {idx}: {temp_file}")
    return audio_bits


audio_files = make_speech_from_chunks(text_pieces)
```

```python
from pydub import AudioSegment
import os

def stitch_audio_files(audio_files, output_path="full_audiobook.mp3"):
    combined_sound = AudioSegment.empty()
    for audio in audio_files:
        sound_clip = AudioSegment.from_mp3(audio)
        combined_sound += sound_clip
        os.remove(audio)  # Clean up the temp file
    combined_sound.export(output_path, format="mp3")
    print(f"All done! Audiobook saved as {output_path}")
    return output_path


audio_files = [f"temp_{i}.mp3" for i in range(183)]


existing_files = [f for f in audio_files if os.path.exists(f)]
if len(existing_files) == 183:
    audiobook = stitch_audio_files(existing_files)
    print("Merging complete! Download full_audiobook.mp3 from /content/")
else:
    print(f"Oops, found {len(existing_files)} files instead of 183. Missing files: {set(audio_files) - set(existing_files)}")
```

```python

full_audio_wav = "/content/full_audiobook.mp3"
sound = AudioSegment.from_mp3(audiobook)
sound.export(full_audio_wav, format="wav")
print(f"Converted {audiobook} to {full_audio_wav}")
```
